# Linear + sparsemax

```python
model_id     = "google/switch-base-16"
dataset_id   = "samsum"
batch_size   = 1
max_src_len  = 128   # max tokens for dialogue
max_tgt_len  = 64    # max tokens for summary
```

```python
import torch
import torch.nn as nn
from torch.profiler import profile, ProfilerActivity
from datasets import load_dataset
from transformers import (
    AutoTokenizer,
    SwitchTransformersForConditionalGeneration,
)
import subprocess, sys
```

```python
# sparsemax
class SparsemaxNorm(nn.Module):
    def __init__(self, dim: int = -1):
        super().__init__()
        self.dim = dim

    def forward(self, input: torch.Tensor) -> torch.Tensor:
        input_shifted = input - input.max(dim=self.dim, keepdim=True).values
        sorted_input, _ = input_shifted.sort(dim=self.dim, descending=True)
        cumsum = sorted_input.cumsum(dim=self.dim)
        range_vals = torch.arange(1, input.size(self.dim) + 1,
                                  device=input.device, dtype=input.dtype)
        shape = [1] * input.dim(); shape[self.dim] = -1
        range_vals = range_vals.view(shape)
        support = (1 + range_vals * sorted_input) > cumsum
        k = support.sum(dim=self.dim, keepdim=True)
        tau = (cumsum.gather(self.dim, k - 1) - 1) / k
        return torch.clamp(input_shifted - tau, min=0.0)
```

```python
class LinearSparsemaxRouter(nn.Module):
    def __init__(self, config):
        super().__init__()
        self.gate = nn.Linear(config.hidden_size, config.num_experts)
        self.norm = SparsemaxNorm(dim=-1)

    def forward(self, hidden_states: torch.Tensor):
        logits = self.gate(hidden_states)                 # [B, T, E]
        probs  = self.norm(logits)                        # [B, T, E]
        z_loss = torch.logsumexp(logits, -1).pow(2).mean()
        return probs, z_loss
```

```python
from transformers.models.switch_transformers import modeling_switch_transformers as mst

def sparsemax_moe_forward(self, hidden_states: torch.Tensor):
    B, T, H = hidden_states.shape
    router_probs, _ = self.router(hidden_states)      # [B, T, E]
    BT = B * T
    hidden_flat = hidden_states.view(BT, H)           # [BT, H]
    router_flat = router_probs.view(BT, -1)           # [BT, E]
    combined_flat = torch.zeros_like(hidden_flat)     # [BT, H]
    num_experts   = len(self.experts)

    for exp_id in range(num_experts):
        w_exp = router_flat[:, exp_id]                # [BT]
        nonzero = w_exp > 0
        if not nonzero.any():
            continue
        idx = nonzero.nonzero(as_tuple=False).squeeze(1)   # [N]
        inputs = hidden_flat[idx]                         # [N, H]
        expert = (self.experts[exp_id]
                  if isinstance(self.experts, nn.ModuleList)
                  else list(self.experts.values())[exp_id])
        out = expert(inputs.unsqueeze(1)).squeeze(1)       # [N, H]
        w_sel = w_exp[idx].unsqueeze(1)                    # [N, 1]
        combined_flat[idx] += w_sel * out

    combined = combined_flat.view(B, T, H)
    dummy_idx = torch.zeros((B, T), dtype=torch.long,
                            device=hidden_states.device)
    return combined, (router_probs, dummy_idx)

mst.SwitchTransformersSparseMLP.forward = sparsemax_moe_forward
```

```python
raw = load_dataset(dataset_id, split="validation").shuffle(seed=42).select(range(batch_size))

# Initialize tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_id)

# Tokenize source dialogues
model_inputs = tokenizer(
    ["summarize: " + d for d in raw["dialogue"]],
    max_length=max_src_len,
    truncation=True,
    padding="max_length",
    return_tensors="pt"
)

# Tokenize target summaries as labels
with tokenizer.as_target_tokenizer():
    targets = tokenizer(
        raw["summary"],
        max_length=max_tgt_len,
        truncation=True,
        padding="max_length",
        return_tensors="pt"
    )
model_inputs["labels"] = targets["input_ids"]

# Move inputs & labels to device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model_inputs = {k: v.to(device) for k, v in model_inputs.items()}
```

```python
model = SwitchTransformersForConditionalGeneration.from_pretrained(model_id)
for module in model.modules():
    if isinstance(module, mst.SwitchTransformersSparseMLP):
        module.router = LinearSparsemaxRouter(model.config)
model.to(device).eval()
```

```python
with profile(
    activities=[ProfilerActivity.CPU, ProfilerActivity.CUDA],
    record_shapes=True,
    profile_memory=True,
    with_flops=True,
) as prof:
    with torch.no_grad():
        # forward pass with real data (includes loss compute)
        outputs = model(**model_inputs)
        _ = outputs.logits    # or use outputs.loss to include loss
```

```python
events = prof.key_averages()
total_flops = sum(evt.flops for evt in events if getattr(evt, "flops", None) is not None)
print(f"Estimated FLOPs for one SAMSum batch forward: {total_flops:.2e}")
```

# linear + topk

```python
model_id     = "google/switch-base-16"
dataset_id   = "samsum"
batch_size   = 1
max_src_len  = 128   # adjust to fit your GPU
max_tgt_len  = 64
```

```python
import torch
import torch.nn as nn
from torch.profiler import profile, ProfilerActivity
from datasets import load_dataset
from transformers import (
    AutoTokenizer,
    SwitchTransformersForConditionalGeneration,
)
import subprocess, sys
```

```python
class SoftmaxNorm(nn.Module):
    def __init__(self, dim: int = -1):
        super().__init__()
        self.dim = dim

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return torch.softmax(x, dim=self.dim)
```

```python
class LinearRouter(nn.Module):
    def __init__(self, config, top_k: int):
        super().__init__()
        self.gate  = nn.Linear(config.hidden_size, config.num_experts)
        self.norm  = SoftmaxNorm(dim=-1)
        self.top_k = top_k

    def forward(self, hidden_states: torch.Tensor):
        # hidden_states: [B, T, H]
        logits       = self.gate(hidden_states)                    # [B, T, E]
        probs        = self.norm(logits)                           # [B, T, E]
        topk_probs, topk_idx = torch.topk(probs, k=self.top_k, dim=-1)
        mask         = torch.zeros_like(probs).scatter_(-1, topk_idx, topk_probs)
        z_loss       = torch.logsumexp(logits, -1).pow(2).mean()   # dummy load‐balance
        return mask, z_loss, topk_idx
```

```python
from transformers.models.switch_transformers import modeling_switch_transformers as mst

def sparsemlp_forward(self, hidden_states: torch.Tensor):
    B, T, H = hidden_states.shape
    router_mask, z_loss, topk_idx = self.router(hidden_states)

    BT         = B * T
    hidden_flat = hidden_states.view(BT, H)    # [BT, H]
    mask_flat   = router_mask.view(BT, -1)     # [BT, E]
    idx_flat    = topk_idx.view(BT, -1)        # [BT, top_k]

    combined_flat = torch.zeros_like(hidden_flat)
    num_experts   = len(self.experts)

    for exp_id in range(num_experts):
        sel   = (idx_flat == exp_id)
        if not sel.any():
            continue
        pos   = sel.nonzero(as_tuple=False)
        bt_idx, slot_idx = pos[:,0], pos[:,1]

        weights = mask_flat[bt_idx, exp_id]    # [N]
        inputs  = hidden_flat[bt_idx]          # [N, H]

        expert = ( self.experts[exp_id]
                  if isinstance(self.experts, nn.ModuleList)
                  else list(self.experts.values())[exp_id] )
        out     = expert(inputs.unsqueeze(1)).squeeze(1)  # [N, H]
        combined_flat[bt_idx] += weights.unsqueeze(1) * out

    combined = combined_flat.view(B, T, H)
    dummy_idx = torch.zeros((B, T), dtype=torch.long,
                            device=hidden_states.device)
    return combined, (router_mask, dummy_idx)

mst.SwitchTransformersSparseMLP.forward = sparsemlp_forward
```

```python
raw = load_dataset(dataset_id, split="validation") \
        .shuffle(seed=42).select(range(batch_size))

tokenizer = AutoTokenizer.from_pretrained(model_id)

# tokenize source
model_inputs = tokenizer(
    ["summarize: " + d for d in raw["dialogue"]],
    max_length=max_src_len,
    truncation=True,
    padding="max_length",
    return_tensors="pt"
)
# tokenize target summaries
with tokenizer.as_target_tokenizer():
    tgt = tokenizer(
        raw["summary"],
        max_length=max_tgt_len,
        truncation=True,
        padding="max_length",
        return_tensors="pt"
    )
model_inputs["labels"] = tgt["input_ids"]

# move to device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model_inputs = {k: v.to(device) for k, v in model_inputs.items()}
```

```python
flops_by_k = {}
for k in range(1, 17):
    # load fresh model
    model = SwitchTransformersForConditionalGeneration.from_pretrained(model_id)
    # inject top-k router
    for module in model.modules():
        if isinstance(module, mst.SwitchTransformersSparseMLP):
            module.router = LinearRouter(model.config, k)
    model.to(device).eval()

    # profile forward+loss
    with profile(
        activities=[ProfilerActivity.CPU, ProfilerActivity.CUDA],
        record_shapes=True,
        profile_memory=True,
        with_flops=True,
    ) as prof:
        with torch.no_grad():
            outputs = model(**model_inputs)
            _ = outputs.loss

    events     = prof.key_averages()
    total_flops = sum(evt.flops for evt in events if getattr(evt, "flops", None) is not None)
    flops_by_k[k] = total_flops
    print(f"top_k={k}: {total_flops:.2e} FLOPs")
```

```python
print("\nFLOPs by top_k:")
for k, fl in flops_by_k.items():
    print(f"  k={k:<2} → {fl:.2e}")
```

# attention

```python
model_id     = "google/switch-base-16"
dataset_id   = "samsum"
batch_size   = 1
max_src_len  = 128   # adjust to fit your GPU
max_tgt_len  = 64
```

```python
import torch
import torch.nn as nn
from torch.profiler import profile, ProfilerActivity
from datasets import load_dataset
from transformers import (
    AutoTokenizer,
    SwitchTransformersForConditionalGeneration,
)
import subprocess, sys
```

```python
class SoftmaxNorm(nn.Module):
    def __init__(self, dim: int = -1):
        super().__init__()
        self.dim = dim
    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return torch.softmax(x, dim=self.dim)

class SparsemaxNorm(nn.Module):
    def __init__(self, dim: int = -1):
        super().__init__()
        self.dim = dim
    def forward(self, input: torch.Tensor) -> torch.Tensor:
        input_shifted = input - input.max(dim=self.dim, keepdim=True).values
        sorted_input, _ = input_shifted.sort(dim=self.dim, descending=True)
        cumsum = sorted_input.cumsum(dim=self.dim)
        range_vals = torch.arange(1, input.size(self.dim) + 1,
                                  device=input.device, dtype=input.dtype)
        shape = [1] * input.dim(); shape[self.dim] = -1
        range_vals = range_vals.view(shape)
        support = (1 + range_vals * sorted_input) > cumsum
        k = support.sum(dim=self.dim, keepdim=True)
        tau = (cumsum.gather(self.dim, k - 1) - 1) / k
        return torch.clamp(input_shifted - tau, min=0.0)
```

```python
class AttentionRouter(nn.Module):
    def __init__(self, config, norm: nn.Module, top_k: int = None):
        super().__init__()
        self.query       = nn.Linear(config.hidden_size, config.hidden_size)
        self.expert_keys = nn.Parameter(torch.randn(config.num_experts, config.hidden_size))
        nn.init.normal_(self.expert_keys, mean=0., std=0.02)
        self.norm  = norm
        self.top_k = top_k if top_k is not None else config.num_experts

    def forward(self, hidden_states: torch.Tensor):
        B, T, H = hidden_states.shape
        q      = self.query(hidden_states).view(-1, H)       # [B*T, H]
        logits = (q @ self.expert_keys.T).view(B, T, -1)     # [B, T, E]
        probs  = self.norm(logits)                           # [B, T, E]
        if self.top_k < probs.size(-1):
            topk_probs, topk_idx = torch.topk(probs, k=self.top_k, dim=-1)
            sparse = torch.zeros_like(probs)
            sparse.scatter_(-1, topk_idx, topk_probs)
            probs = sparse
        z_loss = torch.logsumexp(logits, -1).pow(2).mean()
        return probs, z_loss
```

```python
from transformers.models.switch_transformers import modeling_switch_transformers as mst

def attention_moe_forward(self, hidden_states: torch.Tensor):
    B, T, H     = hidden_states.shape
    router_probs, _ = self.router(hidden_states)      # [B, T, E]
    BT           = B * T
    hidden_flat  = hidden_states.view(BT, H)          # [BT, H]
    router_flat  = router_probs.view(BT, -1)          # [BT, E]
    combined_flat = torch.zeros_like(hidden_flat)     # [BT, H]
    num_experts  = len(self.experts)

    for exp_id in range(num_experts):
        w = router_flat[:, exp_id]                    # [BT]
        nz = w > 0
        if not nz.any(): continue
        idx    = nz.nonzero(as_tuple=False).squeeze(1) # [N]
        inp    = hidden_flat[idx]                     # [N, H]
        expert = (self.experts[exp_id]
                  if isinstance(self.experts, nn.ModuleList)
                  else list(self.experts.values())[exp_id])
        out = expert(inp.unsqueeze(1)).squeeze(1)      # [N, H]
        combined_flat[idx] += w[idx].unsqueeze(1) * out

    combined = combined_flat.view(B, T, H)
    dummy_idx = torch.zeros((B, T), dtype=torch.long, device=hidden_states.device)
    return combined, (router_probs, dummy_idx)

mst.SwitchTransformersSparseMLP.forward = attention_moe_forward
```

```python
raw = load_dataset(dataset_id, split="validation").shuffle(seed=42).select(range(batch_size))
tokenizer = AutoTokenizer.from_pretrained(model_id)

# tokenize source
model_inputs = tokenizer(
    ["summarize: " + d for d in raw["dialogue"]],
    max_length=max_src_len,
    truncation=True,
    padding="max_length",
    return_tensors="pt"
)
# tokenize target (labels)
with tokenizer.as_target_tokenizer():
    tgt = tokenizer(
        raw["summary"],
        max_length=max_tgt_len,
        truncation=True,
        padding="max_length",
        return_tensors="pt"
    )
model_inputs["labels"] = tgt["input_ids"]

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model_inputs = {k: v.to(device) for k, v in model_inputs.items()}
```

```python
flops_softmax = {}
for k in range(1, 17):
    model = SwitchTransformersForConditionalGeneration.from_pretrained(model_id)
    # inject attention-router with softmax + top-k
    for m in model.modules():
        if isinstance(m, mst.SwitchTransformersSparseMLP):
            m.router = AttentionRouter(model.config, SoftmaxNorm(dim=-1), top_k=k)
    model.to(device).eval()

    with profile(
        activities=[ProfilerActivity.CPU, ProfilerActivity.CUDA],
        record_shapes=True,
        profile_memory=True,
        with_flops=True,
    ) as prof:
        with torch.no_grad():
            outputs = model(**model_inputs)
            _ = outputs.loss
    total = sum(evt.flops for evt in prof.key_averages() if getattr(evt, "flops", None) is not None)
    flops_softmax[k] = total
```

```python
model = SwitchTransformersForConditionalGeneration.from_pretrained(model_id)
for m in model.modules():
    if isinstance(m, mst.SwitchTransformersSparseMLP):
        m.router = AttentionRouter(model.config, SparsemaxNorm(dim=-1), top_k=None)
model.to(device).eval()

with profile(
    activities=[ProfilerActivity.CPU, ProfilerActivity.CUDA],
    record_shapes=True,
    profile_memory=True,
    with_flops=True,
) as prof:
    with torch.no_grad():
        outputs = model(**model_inputs)
        _ = outputs.loss
flops_sparsemax = sum(evt.flops for evt in prof.key_averages() if getattr(evt, "flops", None) is not None)
```

```python
print("FLOPs (Attention + Softmax + Top-k):")
for k, f in flops_softmax.items():
    print(f"  k={k:2d} → {f:.2e}")

print(f"\nFLOPs (Attention + Sparsemax): {flops_sparsemax:.2e}")
```
